```python
5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number
```

```python
num=int(input("Enter the number:"))
if(num > 1):
    for i in range(2,num):
        if(num % i == 0):
            print(num,"is not a prime number")
            break;
    else:
        print(num,"is a prime number")
else:
    print(num,"is not a prime number")
```
