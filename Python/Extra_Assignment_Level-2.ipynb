```python
#Print odd number
number=(20,10,16,19,25,1,276,188)
print("(20,10,16,19,25,1,276,188)")
for val in number:
    if(val %2 != 0):
        print(val,"is Odd")
```

```python
#Print even number
number=(20,10,16,19,25,1,276,188)
print("(20,10,16,19,25,1,276,188)")
for val in number:
    if(val %2 == 0):
        print(val,"is Even")
```
