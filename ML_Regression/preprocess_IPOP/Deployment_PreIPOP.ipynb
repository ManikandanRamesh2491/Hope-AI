```python
#Deployment phase
import pickle
import pandas as pd
```

```python
#Use the pandas library funtion read the csv data file and store it in the dataset variable
dataset=pd.read_csv("50_Startups.csv")
```

```python
#Using the get_dummies function, we convert a categorical (nominal) column like “state” into multiple separate columns. This process is called one-hot encoding.
# True=1 and false=0
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True) # remove the dummy column.Here remove the state_California
```

```python
#input - In the dataset column value and store it the independent variable. Only input heading value put here
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
```

```python
#output - In the dataset column value and store it the dependent variable. Only output heading value put here
dependent=dataset[["Profit"]]
```

```python
#Split training and test set using sklearn library and import only train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# Here we can used standardisation in sklearn library and import StandardScalar fumction
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
#standard deviation and mean calculation from x_train value will take same value tranform to x_test
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
scy=StandardScaler()
#standard deviation and mean calculation from preprocess y_train value will take same value tranform to y_test
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
```
