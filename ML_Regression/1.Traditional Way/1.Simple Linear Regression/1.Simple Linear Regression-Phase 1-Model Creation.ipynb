```python
#import the pandas library function in pd
import pandas as pd
```

```python
#Use the pandas library funtion read the csv data file and store it in the dataset variable
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#input - In the dataset YearsExperience column value and store it the independent variable
independent = dataset[["YearsExperience"]] 
```

```python
#output - In the dataset Salary column value and store it the dependent variable
dependent = dataset[["Salary"]]
```

```python
#Split training and test set using sklearn library and import only train_test_split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model creation
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#Calculate weight(w or m value)
weight = regressor.coef_
weight
```

```python
#Calculate bias(c or b value)
bias = regressor.intercept_
bias
```

```python
#Evaluation metrics - process of prediction
y_pred=regressor.predict(x_test)
```

```python
#Calculate the r_Score value and weather check the model is good or bad.
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
#Print the r_score value
r_score
```

```python
#Save the model through pickle library files
import pickle
filename="Finalized_model_linear.sav"
pickle.dump(regressor,open(filename,'wb'))
```

```python
#load the model
loaded_model = pickle.load(open(filename,'rb'))
result=loaded_model.predict([[13]])
                           
```
